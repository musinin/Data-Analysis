```python
# HTML     HTML  1
response = request.urlopen("http://developers.google.com")
bhtml = response.read()
print(bhtml)
chtml = bhtml.decode("utf-8")
#print(chtml)
print(type(chtml))
```

```python
# HTML     HTML  2
response2 = requests.get("http://developers.google.com")
print(response2.content)
chtml2 = response2.content.decode("utf-8")
print(chtml2)
```

```python
# HTML     HTML  3
response3 = requests.get("http://www.naver.com")
print(response3.content)
chtml3 = response3.content.decode("utf-8")
# print(chtml3)
with open('naver1.html','wt') as f:
    f.write(chtml3)
```

```python
# HTML     HTML  4 (   )

header_info ={
    "user-Agent" :"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/134.0.0.0 Safari/537.36" 
}
response4 = requests.get("http://www.naver.com", headers=header_info)
print(response4.content)
chtml4 = response4.content.decode("utf-8")
#print(len(chtml4))
with open('naver2.html','wt') as f:
    f.write(chtml4)
```
